```python
from matplotlib.pyplot import axes


fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(15,10))

sns.barplot(data=AKS, y='value', x='method', hue='score', ax=ax[0][0])
ax[0][0].axhline(0, color='black', linewidth=0.5)
ax[0][0].set_title('Amazon Kindle Store')

sns.barplot(data=ADM, y='value', x='method', hue='score', ax=ax[0][1])
ax[0][1].axhline(0, color='black', linewidth=0.5)
ax[0][1].set_title('Amazon Digital Music')

sns.barplot(data=P10, y='value', x='method', hue='score', ax=ax[1][0])
ax[1][0].axhline(0, color='black', linewidth=0.5)
ax[1][0].set_title('Palco 2010')

sns.barplot(data=MLE, y='value', x='method', hue='score', ax=ax[1][1])
ax[1][1].axhline(0, color='black', linewidth=0.5)
ax[1][1].set_title('Movielens')

# plt.suptitle('Results Summary', y = 0.99, fontsize=14)

fig.tight_layout(h_pad = 2, w_pad = 2)
plt.savefig('images/result_summary.png')
```
